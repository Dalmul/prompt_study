### 사용자의 프롬프트 작성유형

#### 1. 명령/청유형
 - 업장에 산학 실습을 가는데 실습개요와 목표 알려줘
 - 우아한 거짓말 독후가 써줘
 - 이전에 없던 창업아이디어 알려주시겠어요 ?

 **실습**


 - 내용 :
    
    중요한 회의 일정을 변경해야 하는 상황입니다. 팀원에게 보낼 정중하고 간결한 이메일을 작성하는 프롬프트 제작해보세요.

 - 예시 :
    
    중요한 회의 일정을 변경해야해. 팀원들에게 보내는 정중하고 간결한 이메일을 작성해.

    다음 사항을 반드시 포함해.

    1. 회의 일정 변경 사유:  부득이한 개인 사정으로 연차

    2. 변경된 새로운 회의 일정 알림 : 8/7일 오후 2시에서 8/12일 오후 2시, 3A 회의룸으로 변경

    3. 회의 참여를 독려하는 문구 추가

    4. 회의 준비 사항

    5. 마무리 인사말


    ***메일 내용에 반드시 포함해야할 사항을 순서대로, 정확히 기재한다***





#### 2. 역할 지정
 - 너는 고객서비스 상담사처럼 행동해. 고객의 감정에 최대한 공감할 수 있는 표현을 사용해. 아래 [단계]의 순서에 맞게 답변을 생성해줘.
 - 당신은 파이썬을 잘하는 프로그래머 입니다. 만약 내가 코드 만들어줘라고 하면 코드를 만들어 주면 됩니다.

*실습*

    - 내용 :
    당신은 신생 친환경 스타트업의 마케팅 책임자입니다.Z세대를 타겟으로 한 소셜 미디어 마케팅 전략 5가지를 제안하는 프롬프트를 작성해보세요.

    - 예시 :
    당신은 신생 친환경 스타트업의 마케팅 책임자입니다.
    Z세대를 타겟으로 한 소셜 미디어 마케팅 전략 5가지를 제안해주세요.

    <인플루언서 협업> 을 소재로 해주세요.

    - 각 전략의 구체적인 설명
    - 전략이 효과적인 이유
    - 실행 방법
    - 기대되는 결과
    - 성공 여부를 평가하는 방법

***소재를 명확하게 주어줌으로써 LLM이 좋은 결과를 유도***


#### 3. 상황극형
 - 너는 지금 돈이 2만원 밖에 없어. 그런데 번역 레포트를 제출해야하는데, 전문가를 구할 돈이 없어. 너가 대신 내 번역물을 프루프 리딩을 해줘,
 - 우리는 지금 심리상담 센터에 같이 있어. 너는 인공지능이야. 말 시작할 때 '흐음'으로 항상 시작하면서 심리상담을 해주길 바래.

 ***실습***

 - 내용 :
    당신은 중소기업 인사팀장입니다. 우수한 인재를 채용하려고 하지만, 후보자가 요구하는 연봉이 회사 예산을 초과합니다. 이 상황에서 협상 대화를 시뮬레이션하는 프롬프트를 작성해보세요.

 - 예시 :
    지금 우리는 중소기업 의료기기판매 회사의 면접장에 있어.
    후보자가 높은 연봉을 요구해. 나는 이 후보자 마음에 들어. 연봉 협상을 해야하는데, 대화를 시뮬레이션해보면 좋겠어. 어떤 시뮬레이션을 하면 좋을까?

    + 추가 예시 1:

    시나리오 4에 ㅐ해서, 대화 시뮬레이션을 작성해줘.
    \- 후보자:
    \- 면접간:

    분량 : two paragraph

    ***llm은 토큰 단위이기 때문에 문자로 제한하지 말 것 ! 2문장 이런식으로 줄 것***

    + 추가 예시 2:

    좋아 잘했어.
    
    \>\> 이번에는, 후보자와 예상 시나리오도 면접을 미리 연습하고 싶어.후보자가 할 말을 한 턴씩 나눠줄래?

    \-Resonse Format:
    01. 후보자 : 연봉도 중요하지만, 저는 회사의 직원 복지에 더 관심이 많습니다.
    02. 면접자 : __________ (my turn)

    ***좋아 잘했어와 같은 문구로 llm의 답변을 고도화할 수 있다.***
    ***>> 구분자는 앞의 내용과 뒤의 내용이 달라짐을 의미***
    ***위와 같이 예시를 적용하는 것을 Few-Shot Prompting이라고 한다.***
